```python
from math import sqrt, pi
import matplotlib.pyplot as plt

def solveForFrequency(voltage):
    return sqrt( voltage /(4 * (pi**2 * (5 * 10**-9) * (1.6 * 10**-19))))

                
'''
    solving from 1c our maximun v = 158mV
'''
                
                
volt = [i for i in range(159)]
freq = [solveForFrequency(j) for j in range(159)]


plt.xlabel('voltage (V)')
plt.ylabel('frequency (Hz)')
plt.title('the solution to question 1e')
plt.plot(volt, freq)
plt.show()
```
